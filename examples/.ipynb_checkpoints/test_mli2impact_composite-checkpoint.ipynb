```python
import pImpactR as impact
import numpy as np
import matplotlib.pyplot as plt
MLI = impact.MLI
mli = impact.data.dictClass()
pip = impact.data.dictClass()
```

```python
# mass = impact.util.Me
# q = 1
# g0 = 133.537
# b0 = np.sqrt((g0+1.0)*(g0-1.0))/g0
# bg0 = g0*b0
# brho = bg0*mass/impact.util.cLight
# energy = mass*g0

mass = 0.000511005*1.0e9
energy = 1.0e8 + mass
q = -1.0
```

```python
mli.beam = MLI.getElem.beam(mass=mass*1.0e-9,charge=q,energy=energy*1.0e-9)
mli.elemList = MLI.getElem.elemList(mli.beam)
mli.d1= MLI.getElem.drift(name='d1',l=0.4)
mli.d2= MLI.getElem.drift(name='d2',l=0.12)
mli.q1= MLI.getElem.quadrupole(name='q1',l=0.01,k1=10.0)
mli.q2= MLI.getElem.quadrupole(name='q2',l=0.01,k1=-10.0)
mli.b1= MLI.getElem.sbend(name='b1',l=0.81,angle=1.0,fint=0.5,hgap=0.058)
mli.b2= MLI.getElem.sbend(name='b2',l=0.45,angle=0.5,fint=0.5,hgap=0.058)
mli.elemList = mli.elemList + [mli.d1,mli.d2,mli.q1,mli.q2,mli.b1,mli.b2]
mli.ptm = MLI.getElem.ptm()
mli.elemList.append(mli.ptm)
mli.lineList = [MLI.getElem.line(elemList=[mli.d2,mli.b1,mli.d1,mli.q1,mli.d1,mli.b2,mli.d2])]
mli.labor = [mli.lineList[0].name,mli.ptm.name]
MLI.writeInputfile(mli.elemList,mli.lineList,mli.labor)
MLI.run()
M,G=MLI.readTransferMap()
M
```

```python
pip.beam = impact.mli2impact.beam(mli.elemList)
pip.lattice = impact.mli2impact.lattice(mli.elemList,mli.lineList[0])
impact.getTransferMap(pip.beam,pip.lattice,epsilon=[1.0e-9]*6)
```
